```
#sesudah clean data
import pandas as pd
from pandas import DataFrame
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```
data=pd.read_csv('/content/cost.csv')
data.head()
```

```
data.describe()
```

```
 #pisahkan data, buat datframe baru
 X = DataFrame(data,columns=['production_budget_usd'])
 Y = DataFrame(data,columns=['worldwide_gross_usd'])

 
```

```
plt.figure(figsize=(10,10)) #ukuran
plt.scatter(X,Y,alpha=0.3) #alpha buat transparan efek
plt.title('Biaya Film vs keuntungan')
plt.xlabel('budget')
plt.ylabel('keuntungan')
plt.ylim(0,3000000000) #limit
plt.xlim (0,450000000)
plt.show()
```

```
#menjalankan regresi
regresi= LinearRegression()
regresi.fit(X,Y)
```

coef m pada persamaan y=mx+c


```
#melihat m 
regresi.coef_
```

mencari intercept atau c


```
regresi.intercept_
```

```
plt.figure(figsize=(10,10)) #ukuran
plt.scatter(X,Y,alpha=0.3) #alpha buat transparan efek
plt.plot(X,regresi.predict(X), color='red', linewidth=4)  #menggambar regresi linear dengan regresi.predict(X)


plt.title('Biaya Film vs keuntungan')
plt.xlabel('budget')
plt.ylabel('keuntungan')
plt.ylim(0,3000000000) #limit
plt.xlim (0,450000000)
plt.show()
```

```
#periksa apakah sudah bagus model kita pake R, 
#R adalah jarak antara titik dan titik garis prediksi
#persamaan regresi linear sudah dapat, y=mx+c, dengan m atau 3.1 (regresi.coef_) 
#dan C -=7.236.192,7
#cara prediksi untung dengan bajet 50 juta yaitu masukin aja ke persamaan kita, x menjadi 50 juta

#hitung seberapa bagus model kita, model kita 55 persen bener.

regresi.score(X,Y)
 
```
